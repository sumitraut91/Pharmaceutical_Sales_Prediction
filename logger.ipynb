```python
# Set the logging configuration
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Create a logger
logger = logging.getLogger('my_logger')

# Handlers
console_handler = logging.StreamHandler()  
file_handler = logging.FileHandler('app.log')  

logger.addHandler(console_handler)
logger.addHandler(file_handler)

# Formatter
formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
console_handler.setFormatter(formatter)
file_handler.setFormatter(formatter)

# Log the project start
logger.info("Pharmaceutical Sales Prediction Project Started.")

# Load and preprocess data
logger.info("Loading and preprocessing data...")
store_df = pd.read_csv("store.csv")
train_df= pd.read_csv("train.csv")
test_df= pd.read_csv("test.csv")

# Create class to find out missing values

class DatasetAnalyser:
    def __init__(self, df):
        self.df = df

    def get_null_values_count(self):
        return self.df.isnull().sum()

    def get_missing_values_percentage(self):
        total_rows = len(self.df)
        null_counts = self.get_null_values_count()
        return (null_counts / total_rows) * 100


# Perform data analysis
logger.info("Performing data analysis...")
analyzer = DatasetAnalyser(store_df)

# Get the count of null values in each column
null_values_count = analyzer.get_null_values_count()

# Get the percentage of missing values in each column
missing_values_percentage = analyzer.get_missing_values_percentage()

print("Null Values Count:")
print(null_values_count)

print("\nMissing Values Percentage:")
print(missing_values_percentage)

# Find the State,School holiday and Open counts from train data
state_holiday_counts = train_df['StateHoliday'].value_counts()
school_holiday_counts = train_df['SchoolHoliday'].value_counts()
open_closed_counts = train_df['Open'].value_counts()

print("*"*50)
print("StateHoliday Counts:")
print(state_holiday_counts)
print("*"*50)

print("*"*50)
print("\nSchoolHoliday Counts:")
print(school_holiday_counts)
print("*"*50)

print("*"*50)
print("\nOpen Counts:")
print(open_closed_counts)
print("*"*50)


class TrainTestDataCleaner:
    """
        A class to clean the dataframe.
    """
    
    def __init__(self):
        pass
    
    def drop_closed_stores(self, df):
        """
            Drop rows where the 'Open' column is 0(closed stores).
        """
        
        try:
            cleaned = df.query("Open == 1")
            return cleaned
        except:
            pass
        
    def convert_to_datatime(self, df):
        """
            Converts the 'Date' column to datetime format.
        """
        
        try:
            df['Date'] = pd.to_datetime(train_df['Date'])
            return df
        except:
            pass
    
    def sort_by_date(self, df):
        """
            Sort the DataFrame by the 'Date' column.
        """
        return df.sort_values(by=["Date"], ascending=False)
    
    def convert_state_holiday_to_str(self, df):
        """
            Converts the 'StateHoliday' column to a string data type.
        """
        df['StateHoliday'] = df['StateHoliday'].astype(str)
        return df
    
    
    
    def transform_date(self, df):
        """
            Transform 'Year', 'Month' and 'Day' from the 'Date' column.
        """
        
        df['Date'] = pd.to_datetime(df['Date'])
        df['Year'] = pd.DatetimeIndex(df['Date']).year
        df['Month'] = pd.DatetimeIndex(df['Date']).month
        df['Day'] = pd.DatetimeIndex(df['Date']).day
        return df


cleaner = TrainTestDataCleaner()
cleaned_train_df = cleaner.transform_date(train_df)
cleaned_train_df = cleaner.sort_by_date(cleaned_train_df)
cleaned_train_df = cleaner.convert_state_holiday_to_str(cleaned_train_df)
cleaned_train_df

merged_df = cleaned_train_df.merge(store_df, on='Store', how='left')


# Visualize insights
logger.info("Visualizing insights...")
# Sales of each Store on monthly basis
store_sales_month = merged_df.pivot_table(index='Month', columns='StoreType', values='Sales', aggfunc='mean')
store_sales_month.columns = [f'Store-{col} Sales' for col in store_sales_month.columns]
store_sales_month.reset_index(inplace=True)

display(store_sales_month)

months = store_sales_month['Month']
store_a_sales = store_sales_month['Store-a Sales']
store_b_sales = store_sales_month['Store-b Sales']
store_c_sales = store_sales_month['Store-c Sales']
store_d_sales = store_sales_month['Store-d Sales']

logger.info("Monthly Sales for Each Store Type...")
plt.figure(figsize=(10, 6))
plt.plot(months, store_a_sales, label='Store A', marker='o')
plt.plot(months, store_b_sales, label='Store B', marker='o')
plt.plot(months, store_c_sales, label='Store C', marker='o')
plt.plot(months, store_d_sales, label='Store D', marker='o')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales for Each Store Type')
plt.legend()
plt.show()

# Check for distribution in both Train and Test dataset
train_promotion_counts = train_df['Promo'].value_counts()
train_promotion_proportions = train_promotion_counts / len(train_df)*100

print("Training Set Promotion Distribution:")
print(train_promotion_counts)
print(train_promotion_proportions)

test_promotion_counts = test_df['Promo'].value_counts()
test_promotion_proportions = test_promotion_counts / len(test_df)*100

print("Test Set Promotion Distribution:")
print(test_promotion_counts)
print(test_promotion_proportions)

logger.info("Distribution of Promo in Train and Test")
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot distribution of promotions in the training set
sns.distplot(train_df['Promo'], ax=axes[0])
axes[0].set_title("Distribution of Promo in Train")
axes[0].set_xlabel("Promotion")
axes[0].set_ylabel("Frequency")

# Plot distribution of promotions in the test set
sns.distplot(test_df['Promo'], ax=axes[1])
axes[1].set_title("Distribution of Promo in Test")
axes[1].set_xlabel("Promotion")
axes[1].set_ylabel("Frequency")
plt.show()

logger.info("Average Sales on School Holiday vs Non-School Holiday")
#Check & compare sales behaviour before, during and after holidays
school_df_agg = merged_df.groupby("SchoolHoliday")["Sales"].mean()
school_df_agg.index = ["School Holiday", "No School Holiday"]
colors = ["blue", "green"]
plt.bar(school_df_agg.index, school_df_agg.values,color= colors)
plt.xlabel("Holiday")
plt.ylabel("Avg Sales")
plt.title("Average Sales on School Holiday vs Non-School Holiday")
plt.show()

logger.info("Sales comparision on State Holiday and Non State Holiday")

# Sales comparision on State Holiday and Non State Holiday
no_state_holiday = merged_df.query("not (StateHoliday != 0 & StateHoliday != '0')")
state_holiday_df = merged_df.query("StateHoliday == 0 | StateHoliday == '0'")

# Create bar plots for sales on state holidays and no state holidays
plt.figure(figsize=(10, 6))

# Bar plot for sales on state holidays
plt.subplot(1, 2, 1)
sns.barplot(x='StoreType', y='Sales', data=state_holiday_df, ci= None)
plt.xlabel("Store Type")
plt.ylabel("Average Sales")
plt.title("Sales on State Holidays")

# Bar plot for sales on no state holidays
plt.subplot(1, 2, 2)
sns.barplot(x='StoreType', y='Sales', data=no_state_holiday,ci= None)
plt.xlabel("Store Type")
plt.ylabel("Average Sales")
plt.title("Sales on No State Holidays")
plt.tight_layout()
plt.show()


logger.info("Average Sales Comparison for Different Holidays")
# comparison between any seasonal (Christmas, Easter etc) purchase behaviours with State Holiday
holiday_comparison = cleaned_train_df.groupby("StateHoliday")["Sales"].mean()
holiday_comparison.index = ["No holiday", "Public holiday", "Easter", "Christmas"]

print("*"*50)
print(holiday_comparison)
print("*"*50)

plt.figure(figsize=(8, 6))
sns.barplot(x=holiday_comparison.index, y=holiday_comparison.values, ci=None)
plt.xlabel("State Holiday type")
plt.ylabel("Average Sales")
plt.title("Average Sales Comparison for Different Holidays")
plt.show()


logger.info("Comparison of Sales and Customers on Days of the Week")
# Comparison of average sales behaviour in each day of week

import calendar
day_comparison = cleaned_train_df.groupby("DayOfWeek").agg({"Sales": "mean", "Customers": "mean"})

day_names = [calendar.day_name[i] for i in range(7)]

plt.figure(figsize=(10, 6))
sns.barplot(x=day_names, y=day_comparison["Sales"], color="pink",label="Sales")
sns.barplot(x=day_names, y=day_comparison["Customers"], color="green", label="Customers")

for index, value in enumerate(day_comparison["Sales"]):
    plt.text(index, value, f'{value:.0f}', ha='center', va='bottom', color='black', fontsize=9)
for index, value in enumerate(day_comparison["Customers"]):
    plt.text(index, value, f'{value:.0f}', ha='center', va='bottom', color='black', fontsize=9)

plt.xlabel("Day of the Week")
plt.ylabel("Average Value")
plt.title("Comparison of Sales and Customers on Days of the Week")
plt.legend()

plt.show()


logger.info("Sales behaviour for Christmas season 2013-2014")
# Sales behaviour for Christmas season 2013-2014
before_index = (cleaned_train_df["Date"] >= pd.to_datetime("2013-12-01")) & (cleaned_train_df["Date"] < pd.to_datetime("2013-12-25"))
during_index = (cleaned_train_df["Date"] >= pd.to_datetime("2013-12-25")) & (cleaned_train_df["Date"] < pd.to_datetime("2013-12-30"))
after_index = (cleaned_train_df["Date"] >= pd.to_datetime("2014-01-01")) & (cleaned_train_df["Date"] < pd.to_datetime("2014-01-30"))

before_Christmas = cleaned_train_df[before_index].groupby("Date").agg({"Sales": "mean"})
during_Christmas = cleaned_train_df[during_index].groupby("Date").agg({"Sales": "mean"})
after_Christmas = cleaned_train_df[after_index].groupby("Date").agg({"Sales": "mean"})

plt.figure(figsize=(10, 6))

sns.lineplot(x=before_Christmas.index, y=before_Christmas["Sales"], label='Before')
sns.lineplot(x=during_Christmas.index, y=during_Christmas["Sales"], label='During')
sns.lineplot(x=after_Christmas.index, y=after_Christmas["Sales"], label='After')

plt.title("Christmas Sales 2013-2014", size=18)
plt.xticks(rotation=15, fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel(xlabel="Date", fontsize=14)
plt.ylabel(ylabel="Avg Sales", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()

logger.info("Sales behaviour for Christmas season 2014-2015")
# Sales behaviour for Christmas season 2014-2015
before_index = (cleaned_train_df["Date"] >= pd.to_datetime("2014-12-01")) & (cleaned_train_df["Date"] < pd.to_datetime("2014-12-25"))
during_index = (cleaned_train_df["Date"] >= pd.to_datetime("2014-12-25")) & (cleaned_train_df["Date"] < pd.to_datetime("2014-12-30"))
after_index = (cleaned_train_df["Date"] >= pd.to_datetime("2015-01-01")) & (cleaned_train_df["Date"] < pd.to_datetime("2015-01-30"))

before_Christmas = cleaned_train_df[before_index].groupby("Date").agg({"Sales": "mean"})
during_Christmas = cleaned_train_df[during_index].groupby("Date").agg({"Sales": "mean"})
after_Christmas = cleaned_train_df[after_index].groupby("Date").agg({"Sales": "mean"})

plt.figure(figsize=(10, 6))

sns.lineplot(x=before_Christmas.index, y=before_Christmas["Sales"], label='Before')
sns.lineplot(x=during_Christmas.index, y=during_Christmas["Sales"], label='During')
sns.lineplot(x=after_Christmas.index, y=after_Christmas["Sales"], label='After')

plt.title("Christmas Sales 2014-2015", size=18)
plt.xticks(rotation=15, fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel(xlabel="Date", fontsize=14)
plt.ylabel(ylabel="Avg Sales", fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.show()


logger.info("Correlation between sales and number of customers")
# Correlation between sales and number of customers
# Calculate the correlation matrix
correlation_matrix = cleaned_train_df[['Sales', 'Customers']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='plasma', fmt=".2f")
plt.title("Correlation Matrix: Sales vs Customers")
plt.show()

# Comparision in between No.of Customers and sales
# Create a scatter plot to compare the number of customers and sales
sns.scatterplot(x='Sales', y='Customers', data=cleaned_train_df, alpha=0.6)
plt.xlabel("Number of Customers")
plt.ylabel("Sales")
plt.title("Scatter Plot: Number of Customers vs Sales")
plt.grid(True)
plt.show()

logger.info("Effect of Promosion on Existing Customers and Sales")
#Effect of Promosion on Existing Customers and Sales
cleaned_train_df['Date'] = pd.to_datetime(cleaned_train_df['Date'])
cleaned_train_df['Month'] = cleaned_train_df['Date'].dt.month

# Separate data into promotion and non-promotion groups
promotion_data = cleaned_train_df[cleaned_train_df['Promo'] == 1]
non_promotion_data = cleaned_train_df[cleaned_train_df['Promo'] == 0]

# Calculate average sales for each group
average_sales_promotion = promotion_data.groupby('Month')['Sales'].mean().reset_index()
average_sales_non_promotion = non_promotion_data.groupby('Month')['Sales'].mean().reset_index()

# Merge average sales for promotion and non-promotion with the month
merged_data = pd.merge(average_sales_promotion, average_sales_non_promotion, on='Month', suffixes=('_Promo', '_Non_Promo'))

logger.info("Calculate percentage increase in sales for each month")
# Calculate percentage increase in sales for each month
merged_data['Percentage Increase'] = ((merged_data['Sales_Promo'] - merged_data['Sales_Non_Promo']) / merged_data['Sales_Non_Promo']) * 100


plt.figure(figsize=(10, 6))
sns.barplot(x='Month', y='Sales_Promo', data=merged_data, color='blue', label='Promotion')
sns.barplot(x='Month', y='Sales_Non_Promo', data=merged_data, color='orange', label='Non-Promotion')
plt.xlabel('Month')
plt.ylabel('Average Sales')
plt.title('Average Sales during Promotion vs Non-Promotion for Each Month')
plt.legend(title='Promotion Status', loc='upper left')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='Month', y='Percentage Increase', data=merged_data)
plt.xlabel('Month')
plt.ylabel('Percentage Increase in Sales')
plt.title('Sales Increase Percentage after Promotion')
plt.grid(True)
plt.show()

logger.info("Calculate average sales for promotion and non-promotion periods")
# Calculate average sales for promotion and non-promotion periods
average_sales_promotion = cleaned_train_df[cleaned_train_df['Promo'] == 1]['Sales'].mean()
average_sales_non_promotion = cleaned_train_df[cleaned_train_df['Promo'] == 0]['Sales'].mean()

plt.figure(figsize=(8, 6))
sns.barplot(x=['Promotion', 'Non-Promotion'], y=[average_sales_promotion, average_sales_non_promotion])
plt.xlabel('Promotion Status')
plt.ylabel('Average Sales')
plt.title('Effect of Promotion on Sales')
plt.show()


logger.info("Trends of customer behavior during store open and closing times")
# Trends of customer behavior during store open and closing times
non_holiday_sales = merged_df[(merged_df["StateHoliday"] == '0') & (merged_df["Open"] == 1)]

# Get unique stores that open on Sundays
sunday_open_stores = non_holiday_sales[non_holiday_sales["DayOfWeek"] == 7]["Store"].unique()

# Filter data for stores open on Sundays
sunday_sales_df = merged_df[merged_df["Store"].isin(sunday_open_stores)]

# Calculate average number of customers for each day of the week
sunday_open_customers = sunday_sales_df.groupby("DayOfWeek")["Customers"].mean()

plt.figure(figsize=(8, 6))
sns.barplot(x=sunday_open_customers.index, y=sunday_open_customers.values)
plt.xlabel("Day of Week")
plt.ylabel("Average Number of Customers")
plt.title("Customer Behavior for Stores Open on Sundays")
plt.xticks(ticks=range(7), labels=["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"])
plt.tight_layout()
plt.show()


logger.info("Comparison of Average Sales on Weekdays and Weekends")
#Stores which are opened on all weekdays. and their effects of sales on weekends
non_holiday_sales = cleaned_train_df[cleaned_train_df['StateHoliday'] == '0']
# sales on weekdays (Monday to Friday)
week_sales = non_holiday_sales[non_holiday_sales["DayOfWeek"].isin([1, 2, 3, 4, 5])]

# stores that are not open on all weekdays
not_open_all_week_days_stores = week_sales[week_sales["Open"] == 0]["Store"].unique()

week_day_stores = store_df[~store_df["Store"].isin(not_open_all_week_days_stores)]

# stores that are not open on all weekdays
non_week_day_stores = store_df[store_df["Store"].isin(not_open_all_week_days_stores)]


print("Stores that are open on all weekdays:")
week_day_stores

week_day_sales = cleaned_train_df[cleaned_train_df["Store"].isin(week_day_stores["Store"].unique())]
weekend_sales = week_day_sales[week_day_sales["DayOfWeek"].isin([6, 7])]

# average sales for weekdays and weekends
average_sales_weekday = week_day_sales[week_day_sales["DayOfWeek"].isin([1, 2, 3, 4, 5])]["Sales"].mean()
average_sales_weekend = weekend_sales[weekend_sales["DayOfWeek"].isin([6, 7])]["Sales"].mean()

print(f"Average Sales on Weekdays: {average_sales_weekday:.2f}")
print(f"Average Sales on Weekends: {average_sales_weekend:.2f}")

plt.figure(figsize=(8, 6))
sns.barplot(x=['Weekdays', 'Weekends'], y=[average_sales_weekday, average_sales_weekend])
plt.xlabel('Day of the Week')
plt.ylabel('Average Sales')
plt.title('Comparison of Average Sales on Weekdays and Weekends')
plt.show()


logger.info("Check how the assortment type affects sales")
#Check how the assortment type affects sales

merged_df['Date'] = pd.to_datetime(merged_df['Date'])

# Extract the month from the 'Date' column
merged_df['Month'] = merged_df['Date'].dt.month

# Filter data for each assortment type
assortment_a_data = merged_df[merged_df['Assortment'] == 'a']
assortment_b_data = merged_df[merged_df['Assortment'] == 'b']
assortment_c_data = merged_df[merged_df['Assortment'] == 'c']

# Group data by month and calculate the average sales for each assortment type
assortment_a_sales = assortment_a_data.groupby('Month')['Sales'].mean()
assortment_b_sales = assortment_b_data.groupby('Month')['Sales'].mean()
assortment_c_sales = assortment_c_data.groupby('Month')['Sales'].mean()

# Create a table to display the results
assortment_table = pd.DataFrame({
                                'A-type assortment sales': assortment_a_sales,
                                'B-type assortment sales': assortment_b_sales,
                                'C-type assortment sales': assortment_c_sales
                                })

assortment_table

plt.figure(figsize=(10, 6))
sns.barplot(data=assortment_table)
plt.xlabel('Assortment Types')
plt.ylabel('Average Sales')
plt.title('Sales by Assortment Types')
plt.show()


logger.info("Correlation between sales and CompetitionDistance")
# Calculate the correlation between sales and CompetitionDistance
correlation = merged_df['Sales'].corr(merged_df['CompetitionDistance'])
print("Correlation between Sales and CompetitionDistance:", correlation)

competition_df = merged_df[["Store", "Sales", "Customers", "StoreType", "CompetitionDistance"]]

competition_df_agg = competition_df.groupby("Store").agg({
                                                            "Sales": "sum",
                                                            "Customers": "sum",
                                                            "CompetitionDistance": "min"
})

plt.figure(figsize=(8, 6))
sns.scatterplot(x='CompetitionDistance', y='Sales', data=competition_df_agg)
plt.xlabel('Competition Distance (meters)')
plt.ylabel('Total Sales')
plt.title('Competition Distance vs Sales')
plt.grid(True)
plt.show()

#correlation coefficient between the minimum competition distance and total sales.
competition_df_corr = merged_df[["Sales", "CompetitionDistance"]].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(competition_df_corr, annot=True, cmap='coolwarm')
plt.title('Correlation between Competition Distance and Sales')
plt.show()

logger.info("Average Sales Before vs After Competition Opening")
#opening or reopening of new competitors affect stores
has_open_competition = merged_df[merged_df['CompetitionOpenSinceMonth'].notnull()]

# Convert the "CompetitionOpenSinceYear" and "CompetitionOpenSinceMonth" columns to datetime format
has_open_competition["CompetitionOpenSinceDate"] = pd.to_datetime(
                                                                    has_open_competition["CompetitionOpenSinceYear"].astype(int).astype(str) + '-' +
                                                                    has_open_competition["CompetitionOpenSinceMonth"].astype(int).astype(str) + '-01'
)

# Function to calculate if a sale occurred before or after the competition opened
def before_or_after_competition(row):
    return row["Date"] < row["CompetitionOpenSinceDate"]

# Apply the function to create the "BeforeCompetition" column
has_open_competition["BeforeCompetition"] = has_open_competition.apply(before_or_after_competition, axis=1)

# Filter the relevant columns for analysis
competition_df = has_open_competition[["Date", "Sales", "Store", "BeforeCompetition"]]

# Calculate the average sales before and after the competition opened
competition_df_agg = competition_df.groupby("BeforeCompetition").agg({"Sales": "mean"})

plt.figure(figsize=(8, 6))
x_col = ["After Competition", "Before Competition"]
sns.barplot(x=x_col, y=competition_df_agg["Sales"])
plt.xlabel('Before or After Competition')
plt.ylabel('Average Sales')
plt.title('Average Sales Before vs After Competition Opening')
plt.show()


logger.info("Yearly,Monthly and Daily Sales Trends of Store Types A, B, C, and D")
#Effect of Percentage drop in sales on Store types after the competition open

# Group the data by "Date" to calculate the total sales for each day
daily_sales = merged_df.groupby('Date')['Sales'].sum().reset_index()

# Plot the total sales on a daily basis using a line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Sales', data=daily_sales)
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Daily Trend of Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Monthly sales trend for each store type
monthly_trend_a = merged_df[merged_df["StoreType"] == "a"].groupby("Month").agg({"Sales": "mean"})
monthly_trend_b = merged_df[merged_df["StoreType"] == "b"].groupby("Month").agg({"Sales": "mean"})
monthly_trend_c = merged_df[merged_df["StoreType"] == "c"].groupby("Month").agg({"Sales": "mean"})
monthly_trend_d = merged_df[merged_df["StoreType"] == "d"].groupby("Month").agg({"Sales": "mean"})

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8), sharex=True)

# Store A
sns.lineplot(x=monthly_trend_a.index, y=monthly_trend_a["Sales"], label='Store A', linewidth=2, ax=axes[0, 0])
axes[0, 0].set_title('Monthly Sales Trend of Store A')

# Store B
sns.lineplot(x=monthly_trend_b.index, y=monthly_trend_b["Sales"], label='Store B', linewidth=2, ax=axes[0, 1])
axes[0, 1].set_title('Monthly Sales Trend of Store B')

# Store C
sns.lineplot(x=monthly_trend_c.index, y=monthly_trend_c["Sales"], label='Store C', linewidth=2, ax=axes[1, 0])
axes[1, 0].set_title('Monthly Sales Trend of Store C')

# Store D
sns.lineplot(x=monthly_trend_d.index, y=monthly_trend_d["Sales"], label='Store D', linewidth=2, ax=axes[1, 1])
axes[1, 1].set_title('Monthly Sales Trend of Store D')

# Adjust layout and labels

plt.xlabel('Month')
plt.ylabel('Average Sales')
plt.suptitle('Monthly Sales Trends of Store Types A, B, C, and D', fontsize=16)
plt.show()

# Yearly sales trend for each store type
yearly_trend_a = merged_df[merged_df["StoreType"] == "a"].groupby("Year").agg({"Sales": "mean"})
yearly_trend_b = merged_df[merged_df["StoreType"] == "b"].groupby("Year").agg({"Sales": "mean"})
yearly_trend_c = merged_df[merged_df["StoreType"] == "c"].groupby("Year").agg({"Sales": "mean"})
yearly_trend_d = merged_df[merged_df["StoreType"] == "d"].groupby("Year").agg({"Sales": "mean"})

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8), sharex=True)

# Store A
sns.lineplot(x=yearly_trend_a.index, y=yearly_trend_a["Sales"], label='Store A', linewidth=2, ax=axes[0, 0])
axes[0, 0].set_title('Yearly Sales Trend of Store A')
axes[0, 0].set_xticks(yearly_trend_a.index)

# Store B
sns.lineplot(x=yearly_trend_b.index, y=yearly_trend_b["Sales"], label='Store B', linewidth=2, ax=axes[0, 1])
axes[0, 1].set_title('Yearly Sales Trend of Store B')
axes[0, 1].set_xticks(yearly_trend_b.index)

# Store C
sns.lineplot(x=yearly_trend_c.index, y=yearly_trend_c["Sales"], label='Store C', linewidth=2, ax=axes[1, 0])
axes[1, 0].set_title('Yearly Sales Trend of Store C')
axes[1, 0].set_xticks(yearly_trend_c.index)

# Store D
sns.lineplot(x=yearly_trend_d.index, y=yearly_trend_d["Sales"], label='Store D', linewidth=2, ax=axes[1, 1])
axes[1, 1].set_title('Yearly Sales Trend of Store D')
axes[1, 1].set_xticks(yearly_trend_d.index)


plt.xlabel('Year')
plt.ylabel('Average Sales')
plt.suptitle('Yearly Sales Trends of Store Types A, B, C, and D', fontsize=16)
plt.tight_layout()
plt.show()


logger.info("Daily Sales Trend for Each Year")
#Daily Sales Trend for Each Year
merged_df['Year'] = merged_df['Date'].dt.year

# Group the data by year and date to calculate total sales for each day within each year
yearly_daily_sales = merged_df.groupby(['Year', 'Date'])['Sales'].sum().reset_index()

# Create a separate line plot for each year
plt.figure(figsize=(12, 6))

for year in yearly_daily_sales['Year'].unique():
    year_data = yearly_daily_sales[yearly_daily_sales['Year'] == year]
    sns.lineplot(x='Date', y='Sales', data=year_data, label=f'Year {year}')

plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.title('Daily Sales Trend for Each Year')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


logger.info("Daily Customers Trend for Each Year")
#Daily Customers Trend for Each Year

merged_df['Year'] = merged_df['Date'].dt.year

# Group the data by year and date to calculate total sales for each day within each year
yearly_daily_sales = merged_df.groupby(['Year', 'Date'])['Customers'].sum().reset_index()

# Create a separate line plot for each year
plt.figure(figsize=(12, 6))

for year in yearly_daily_sales['Year'].unique():
    year_data = yearly_daily_sales[yearly_daily_sales['Year'] == year]
    sns.lineplot(x='Date', y='Customers', data=year_data, label=f'Year {year}')

plt.xlabel('Date')
plt.ylabel('Total Customers')
plt.title('Daily Customers Trend for Each Year')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()



# Concluding remarks
logger.info("to be continued.")
```
